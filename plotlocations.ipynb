```python
import numpy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D 
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
locations = numpy.load("locations.npz")
locs = locations['locations']
```

```python
locs.shape
```

```python
us = locs[:,0]
vs = locs[:,1]
ws = locs[:,2]
fig = plt.figure(figsize=(7,6))
ax = fig.add_subplot(111)
#ax = plt.subplot(111, projection='3d')
s = ax.scatter(us,vs,c=ws,marker='D',alpha=0.8)
ax.set_title('LWA Antenna Locations')
ax.set_xlabel('x'); ax.set_ylabel('y'); #ax.set_zlabel('z')
ax.set_xlim([-100,80]); ax.set_ylim([-90,90]); #ax.set_zlim([-7.5,15.0]);

inset_outrigger = inset_axes(ax, width="30%", height="30%",loc=3,bbox_to_anchor=(0.12,0.1,1,1), bbox_transform=ax.transAxes)
inset_outrigger.scatter(us,vs,c=ws,marker='D',alpha=0.8)
inset_outrigger.set_xlim([-300,-280])
inset_outrigger.set_ylim([0,20])
inset_outrigger.set_xlabel('x'); inset_outrigger.set_ylabel('y');

cbar = fig.colorbar(s, ax=ax)
cbar.set_label('w',rotation=0)
plt.savefig('antenna_locs.pdf')
```
